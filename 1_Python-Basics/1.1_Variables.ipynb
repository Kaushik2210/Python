### Variables in Python

Variables are fundamental elements in programming used to store data that can be referenced and manipulated in a program. In Python, variables are created when you assign a value to them, and they do not need explicit declaration to reserve memory space. The declaration happens automatically when you assign a value to a variable.

```python
##Understanding Variable types
## Python is dynamically typed and is determined at runtime

age=21
height=5.11
is_Student=True

print(type(age)) # <class 'int'>
print(type(height)) # <class 'float'>
```

```python
##Type checking and conversion
age=21
print(type(age)) # <class 'int'>
age_str=str(age) # Convert int to string
print(type(age_str)) # <class 'str'>
```

```python
age='25'
age_int=int(age) # Convert string to int
print(type(age_int)) # <class 'int'>

name="Kaushik"
int(name) # This will raise an error because "Kaushik" cannot be converted to an integer

height=5.11
type(height) # <class 'float'>
int(height) # This will convert the float to an integer by truncating the decimal part, resulting in 5
print(int(height)) # 5
```

```python
##Dynamic Typing
##Python allows the type of variable to change as the program executes

var=10
print(var, type(var)) # 10 <class 'int'>

var="Now I'm a string"
print(var, type(var)) # Now I'm a string <class 'str'>

var=3.14
print(var, type(var)) # 3.14 <class 'float'>
```
