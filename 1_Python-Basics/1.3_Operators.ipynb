```python
## ArithmeticOperation
a=10
b=5
add=a+b
sub=a-b
mul=a*b
div=a/b
floor_div=a//b
mod=a%b
print("Addition: ", add)
print("Subtraction: ", sub)
print("Multiplication: ", mul)
print("Division: ", div)
print("Modulus: ", mod)
print("Floor Division: ", floor_div)
```

```python
## Comparison operators
a=10
b=10
c=5
a==b # True, because values are equal
a!=b # False, because values are equal
a>b # False, because 10 is not greater than 10
a<b # False, because 10 is not less than 10
a>=b # True, because 10 is greater than or equal to 10
a<=b # True, because 10 is less than or equal to 10

str1="Hello"
str2="hello"
str1==str2 # False, because string comparison is case-sensitive
str1!=str2 # True, because "Hello" is not equal to "hello"
a>c # True, because 10 is greater than 5
a<c # False, because 10 is not less than 5
```

```python
## Logical Operators
x=5
y=10
print(x>3 and y<15) # True, because both conditions are true
print(x>3 or y>15) # True, because at least one condition is true
print(not(x>3)) # False, because x>3 is true, and not True
print(not(y>15)) # True, because y>15 is false, and not False
```
